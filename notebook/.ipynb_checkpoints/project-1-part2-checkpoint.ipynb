4 Printout of dataset distribution, including missing data. For imagery datasets, provide the "average image" for each class. For tabular data, use value_counts() and describe(). For textual data, show the distribution of your labels/targets.

````python
import matplotlib.pyplot as plt
import cv2

image_data = []
for img in X_train:
    this_image = cv2.imread(img, 1)
    
    img = cv2.cvtColor(this_image,cv2.COLOR_BGR2RGB)
    border_v = 0
    border_h = 0
    if (1) >= (img.shape[0]/img.shape[1]):
        border_v = int((((1)*img.shape[1])-img.shape[0])/2)
    else:
        border_h = int((((1)*img.shape[0])-img.shape[1])/2)
    img = cv2.copyMakeBorder(img, border_v, border_v, border_h, border_h, cv2.BORDER_REPLICATE, 0)
    img = cv2.resize(img, (160, 160))
    image_data.append(img)
    
avg_image = image_data[0]
for i in range(len(image_data)):
    if i == 0:
        pass
    else:
        alpha = 1.0/(i + 1)
        beta = 1.0 - alpha
        avg_image = cv2.addWeighted(image_data[i], alpha, avg_image, beta, 0.0)

plt.imshow(avg_image)
plt.show()
````

- Currently this dataset has about 1000 anime portraits without background, but the remaining 95% images are with backgrounds. If 1000 portraits aren't enough for training the model. We will also tries to remove the background for the remaining 95% images and feed those images to the model as well.

````python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')

img = cv2.imread("/mnt/c/ML-Project/moeimouto-faces/013_saber/face_54_354_65.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
border_v = 0
border_h = 0
if (1) >= (img.shape[0]/img.shape[1]):
    border_v = int((((1)*img.shape[1])-img.shape[0])/2)
else:
    border_h = int((((1)*img.shape[0])-img.shape[1])/2)
img = cv2.copyMakeBorder(img, border_v, border_v, border_h, border_h, cv2.BORDER_REPLICATE, 0)
img = cv2.resize(img, (160, 160))
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_invert = cv2.bitwise_not(img_gray)



img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
final = cv2.divide(img_gray, 255 - img_smoothing, scale=255)
f = cv2.bitwise_not(final)
(thresh, blackAndWhiteImage) = cv2.threshold(f, 1, 255, cv2.THRESH_BINARY)
plt.figure(figsize=(20,20))
plt.subplot(1,6,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,6,2)
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,6,3)
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,6,4)
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.subplot(1,6,5)
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.subplot(1,6,6)
plt.imshow(blackAndWhiteImage)
plt.axis("off")
plt.title("Final Sketch B/W")
plt.show()
````

11 For tabular data: Use a heatmap to show the correlation between all feature pairs. Discuss, if any, which features you would recommend dropping from your model. Also discuss why you would want to drop them (what is the expected benefit?). For imagery/text: Show a histogram of the distribution of pixels or word embeddings across your dataset.

- In this case, the histogram of the distribution of pixels doesn't provide much information as the color set in dataset is quiet random.
- The histogram of generated hand drawn sketches will result in only 0 and 255, as it has been intended converted to pure black and white. 
- The histogram of original `moeimouto-faces` and black/white `moeimouto-black` dataset will be shown below

````python
import glob
import numpy as np

nb_bins = 256
count_r = np.zeros(nb_bins)
count_g = np.zeros(nb_bins)
count_b = np.zeros(nb_bins)
for image in glob.glob('/mnt/c/ML-Project/moeimouto-faces/000_hatsune_miku/*.png'):    
    x = cv2.imread(image)
    hist_r = np.histogram(x[0], bins=nb_bins, range=[0, 255])
    hist_g = np.histogram(x[1], bins=nb_bins, range=[0, 255])
    hist_b = np.histogram(x[2], bins=nb_bins, range=[0, 255])
    count_r += hist_r[0]
    count_g += hist_g[0]
    count_b += hist_b[0]

bins = hist_r[1]
fig = plt.figure()
plt.bar(bins[:-1], count_r, color='r', alpha=0.33)
plt.bar(bins[:-1], count_g, color='g', alpha=0.33)
plt.bar(bins[:-1], count_b, color='b', alpha=0.33)
````
