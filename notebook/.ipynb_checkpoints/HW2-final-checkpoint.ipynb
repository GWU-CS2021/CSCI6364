```python
import pandas

dataframe = pandas.read_csv('student-mat.csv')
print(dataframe.head())

for c in dataframe.columns:
    print(c)
    print(dataframe[c].value_counts(dropna=False))
print(dataframe.columns)

from sklearn.model_selection import train_test_split

target = 'internet'
y = dataframe[target].map({'yes': 1, 'no': 0})

X = dataframe.drop(labels = [target], axis='columns')


# drop non-numberic fields
yes_no = ['activities','schoolsup','famsup','paid','nursery','higher','romantic']
# map yes/no to 1/0
for c in yes_no:
    X[c] = X[c].map({'yes': 1, 'no': 0})
# translate non_numeric fields to int
non_numeric = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob',
   'Fjob', 'reason', 'guardian']
for c in non_numeric:
    X[c] = X[c].astype('category').cat.codes
```

0. Correctly changes all features of the DataFrame to be numeric, using apply() or other methodology in your notebook (write python code instead of doing it manually or dropping columns like in the first homework)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
# split validate set and test set
validate_train,holdout_train,validate_test,holdout_test = train_test_split(X_test, y_test, test_size=0.5)
```

1. Correctly split dataset into train, validate, and holdout Dataframes

```python
# Import the model we are using
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

baseline_classifier_model = RandomForestClassifier()
# Train the baseline model on training data
baseline_classifier_model = baseline_classifier_model.fit(X_train, y_train)
```

3. A RandomForest model is correctly trained on training data (no validation), and evaluated on the holdout

```python
# Use the forest's predict method on the holdout
predictions = baseline_classifier_model.predict(holdout_train)
# Calculate the accuracy_score
accuracy_score(predictions, holdout_test)
```

4. Accuracy of the baseline RandomForest model is correctly calculated and reported on the holdout

```python
# train the model for 5 times and get the best model for 
best_model = None
classifier_modelt = []
for i in range(5):
    classifier_model = RandomForestClassifier()
    classifier_modelt[i] = classifier_model.fit(X_train, y_train)
    predictions = classifier_modelt[i].predict(validate_train)
    score = accuracy_score(validate_test, predictions)
    print("%d times running, scoring %f"%(i+1,score))
    if best_model is None or best_model[1] < score:
        best_model = (classifier_modelt[i], score)

model, score = best_model
print("best performance %f"%(score))
```

5. The RandomForest model is re-trained five times, and tested on the validation set (reporting best-performing accuracy)

```python
# apply the best model on holdout data
predictions = model.predict(holdout_train)
# Calculate the accuracy_score
accuracy_score(predictions, holdout_test)
```

6. The best performing RandomForest model is correctly tested and scored on the holdout dataset
